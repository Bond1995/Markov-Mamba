```python
import numpy as np
```

```python
r = 1 # ratio n01/n00; n01 = r*n00
n00 = np.array([5, 10, 20, 30, 40, 50]) # number of 00 transitions
b = 0.5 # beta in the add-beta estimator
```

```python
p = 1.0 / (1.0 + np.exp(-0.3656 - 0.0735*n00 + 0.1091*r*n00)) # probability estimated by the model
```

```python
q = 1.0 / (1.0 + (r*n00 + b)/(n00 + b)) # probability estimated by the add-beta estimator
```

```python
print(p)
print(q)
```
